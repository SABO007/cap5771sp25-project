### Dropping columns such as Unnamed 0, Gender, Location

### Renaming the column names

```python

student_data = student_data.rename(columns={
    "Year of Study": "Year",
    "Branch": "Branch",
    "Percentage in Class 12th": "Percentage_12th",
    "How many programming languages do you know": "Programming_Languages",
    "Actively involved in Specific technology": "Specific_Technology",
    "Exposure to Data structure and algorithms": "DSA_Exposure",
    "Exposure to GitHub": "GitHub_Exposure",
    "Are you associated with any developer Community": "Dev_Community",
    "Active in developer Communities": "Active_in_Dev_Community",
    "Participating in Hackathons and other competitions": "Hackathons",
    "Have you made any Software based projects": "Software_Projects",
    "Have you made any Hardware based project (Arduino, Raspberry Pi, Robots or any other)": "Hardware_Projects",
    "Have you ever pitched any idea": "Pitched_Idea"
})

print(student_data)
```

### Declaring Career Goals on the basis of Branch

```python
branch_career_goals = {
    "CSE/IT": ["Data Scientist", "Software Engineer", "Machine Learning Engineer", "Cloud Engineer"],
    "CSIT": ["Data Scientist", "Software Engineer", "Web Developer", "DevOps Engineer"],
    "CSE": ["Software Engineer", "Data Scientist", "Machine Learning Engineer", "Cybersecurity Analyst"],
    "IT": ["Software Engineer", "Web Developer", "Cloud Engineer", "Data Analyst"],
    "CS": ["Software Engineer", "Data Scientist", "Machine Learning Engineer", "AI Engineer"],
    "CSAI": ["AI Engineer", "Machine Learning Engineer", "Data Scientist", "Robotics Engineer"],
    "ECE": ["Embedded Systems Engineer", "IoT Developer", "Signal Processing Engineer", "VLSI Engineer"],
    "EN": ["Power Systems Engineer", "Electrical Design Engineer", "Control Systems Engineer", "Renewable Energy Engineer"],
    "CSME": ["Robotics Engineer", "Automation Engineer", "AI in Manufacturing", "Product Design Engineer"],
    "ME": ["Robotics Engineer", "Automation Engineer", "Thermal Engineer", "Design Engineer"],
    "CE": ["Structural Engineer", "Construction Manager", "Geotechnical Engineer", "Transportation Engineer"],
    "MCA": ["Software Developer", "Data Analyst", "Web Developer", "Database Administrator"]
}
```

### Assigning Careers Goals on the basis of the Branch

```python
def assign_random_career_goal(branch):
    return random.choice(branch_career_goals.get(branch, ["Unknown"]))

student_data["Career_Goal"] = student_data["Branch"].apply(assign_random_career_goal)
```

### Declaring Skills required on the basis of Career Goals

```python
skills_mapping = {
    "Cloud Engineer": ["Cloud Computing", "AWS", "Azure", "Networking", "DevOps", "Virtualization"],
    "Data Scientist": ["Python", "R", "Data Analysis", "Machine Learning", "SQL", "Data Visualization"],
    "Data Analyst": ["SQL", "Excel", "Data Visualization", "Python", "R", "Tableau", "Power BI"],
    "AI Engineer": ["Python", "AI", "Machine Learning", "Neural Networks", "Deep Learning", "TensorFlow"],
    "Machine Learning Engineer": ["Python", "Machine Learning", "Deep Learning", "TensorFlow", "PyTorch", "Algorithms", "Data Structures"],
    "Software Engineer": ["Python", "Java", "C++", "Data Structures", "Algorithms", "Git"],
    "Control Systems Engineer": ["Control Theory", "MATLAB", "Simulink", "Signal Processing", "Embedded Systems"],
    "Web Developer": ["HTML", "CSS", "JavaScript", "React", "Node.js", "Angular", "APIs", "Databases"],
    "Robotics Engineer": ["Robotics", "AI", "C++", "Embedded Systems", "MATLAB", "ROS"],
    "Automation Engineer": ["Automation Systems", "PLC Programming", "Control Systems", "Robotics", "SCADA"],
    "AI in Manufacturing": ["AI", "Manufacturing Systems", "Robotics", "Data Analytics", "Machine Learning"],
    "Thermal Engineer": ["Thermodynamics", "Heat Transfer", "Fluid Mechanics", "MATLAB", "Computational Fluid Dynamics (CFD)"],
    "Design Engineer": ["CAD", "Mechanical Design", "Manufacturing Processes", "Prototyping", "SolidWorks", "AutoCAD"],
    "Structural Engineer": ["Structural Analysis", "CAD", "AutoCAD", "Revit", "Civil Engineering", "Building Codes"],
    "Construction Manager": ["Project Management", "Construction Scheduling", "AutoCAD", "Cost Estimation", "Site Management"],
    "Geotechnical Engineer": ["Soil Mechanics", "Geotechnical Engineering", "Foundation Design", "MATLAB", "AutoCAD"],
    "Transportation Engineer": ["Traffic Analysis", "GIS", "Road Design", "Civil Engineering", "AutoCAD", "Project Management"],
    "Software Developer": ["Java", "C#", "Python", "Data Structures", "Algorithms", "Web Development", "Database Management"],
    "Database Administrator": ["SQL", "Database Management", "Oracle", "MySQL", "Data Security", "Backup and Recovery"]
}

def append_skills(row):
    career_goal = row['Career_Goal']
    if career_goal in skills_mapping:
        return ', '.join(skills_mapping[career_goal]) 
    else:
        return ''

student_data['Skills'] = student_data.apply(append_skills, axis=1)
```

```python
student_data.head()
```

### Removing different naming for unique feature

```python
student_data['Branch'].unique()
```

```python
student_data['Branch'] = student_data['Branch'].replace('CSE', 'CS')
student_data['Branch'] = student_data['Branch'].replace('CSE/IT', 'CSIT')
```

### Scaling Percentage to the median eliminating the range

```python
student_data['Percentage_12th'].unique()
```

```python
avg = {
    '>=90%': 95,
    '>=80% and <90%': 85,
    '>=70% and <80%': 75,
    '>=60% and <70%': 65,
    '<60%': 50

}

student_data['Percentage_12th'] = student_data['Percentage_12th'].map(avg)
```

### Considering only required features for our analysis

```python
final_student_data = student_data[['Branch', 'Percentage_12th', 'Skills','Career_Goal']]
final_student_data
```

### Remove outliers

```python
len(student_data)
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

student_data = remove_outliers(student_data, "Percentage_12th")
```

### Convert job posting date into timestamp

```python
job_data["Job Posting Date"] = pd.to_datetime(job_data["Job Posting Date"], errors="coerce")

job_data["Job Posting Timestamp"] = job_data["Job Posting Date"].astype("int64") // 10**9 
```

### Removal of outliers in job posting date

```python

job_data = remove_outliers(job_data, "Job Posting Timestamp")

job_data["Job Posting Date"] = pd.to_datetime(job_data["Job Posting Timestamp"], unit="s")
job_data.drop(columns=["Job Posting Timestamp"], inplace=True)
```
